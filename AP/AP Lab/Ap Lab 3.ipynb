#### 2. Write a function Projectile () such that it can accept two variables initial velocity and angle and calculate the Maximum Height and Range of projectile. And also it must return both Maximum Height and Range in a single return call. Also plot the graph of Maximum Height and Range with respect to angle from 0 to 360 degree using matplotlib.

```python
def projectile(v1,angle):
    angle = radians(angle)
    Max_height = (v1**2 * np.sin(angle)**2)/(2*g)
    Max_range = (v1**2 * np.sin(2*angle))/g
    return [Max_range,Max_height]
```

```python
g = 9.8
ranges = []
heights = []
v1 = int(input("Enter intial velocity value: "))
for Angle in range(0,360,1) :
    r,h = projectile(v1,Angle)
    ranges.append(r)
    heights.append(h)

plt.plot(np.arange(0,360,1),heights,label='Height')
plt.plot(np.arange(0,360,1),ranges,label='Range')

plt.xlabel("Angle")
plt.ylabel("Height / Range")
plt.title('Projectile Motion With Multiple Angles')
plt.legend()
plt.show()
```

#### 3. Write a function SHM() such that it can accept four variables (amplitude ,angular frequency , phase constant , time ) and return displacement ,velocity ,and acceleration in a single call.
- x = xm Cos (ɷt + φ)
- v = - xm ɷ Sin (ɷt + φ)
- a = xm ɷ 2 Cos (ɷt + φ)

```python
def SHM(amp,freq,phase,time):
    displacement = amp*cos(freq*time + phase)
    velocity = -amp*freq*sin(freq*time + phase)
    acceleration = amp * freq**2 *cos(freq*time + phase)
    return displacement, velocity, acceleration
```

```python
amp = int(input("Enter Amplitude: "))
freq = int(input("Enter frequency: "))
phase = int(input("Enter phase: "))
dis=[]
vel=[]
acc=[]
angle_list = np.arange(0,2*np.pi,0.1)
for time in angle_list:
    d, v, a = SHM(amp,freq,phase,time)
    dis.append(d)
    vel.append(v)
    acc.append(a)
plt.title("Simple Harmonic Motion")
plt.plot(angle_list,dis,label='displacement')
plt.plot(angle_list,vel,label='velocity')
plt.plot(angle_list,acc,label='acceleration')
# plt.grid()
plt.legend()
plt.ylabel("Displacement / Velocity/ Acceleration")
plt.xlabel("Time")
plt.show()
```

#### c) Three vectors a , b and c each have a magnitude of 50 m and lie in an xyplane. Their directions relative to the positive direction of the x axis are 30°, 195°, and 315°, respectively. What are (i) the magnitude and the angle of the vector a+b+c , and (ii) the magnitude and the angle of a- b+c? What are the (iii) magnitude and angle of a fourth vector d such that (a+b ) – (c+d) = 0 ?

```python
angle_a = radians(30)
angle_b = radians(195)
angle_c = radians(315)

Mag = 50

#a
ax= Mag*cos(angle_a)
ay= Mag*sin(angle_a)

#b
bx= Mag*cos(angle_b)
by= Mag*sin(angle_b)

#c
cx= Mag*cos(angle_c) 
cy= Mag*sin(angle_c)
```

```python
# (i) the magnitude and the angle of the vector a+b+c .?
Rx = ax + bx + cx
Ry = ay + by + cy
R = sqrt(Rx**2 + Ry**2)
print(f"Resultant Magnitude = {R}")
angle = atan(Ry/Rx)
if angle < 0:
    angle = 360 + degrees(angle)
print(f"Angle between a+b+c : {angle}")
```

```python
# (ii) the magnitude and the angle of a- b+c?
Rx = ax - bx + cx
Ry = ay - by + cy
R = sqrt(Rx**2 + Ry**2)
print(f"Resultant Magnitude = {R}")
angle = atan(Ry/Rx)
if angle < 0:
    angle = 360 + degrees(angle)
else:
    angle = degrees(angle)
print(f"Angle between a-b+c : {angle}")
```

```python
# (iii) magnitude and angle of a fourth vector d such that (a+b ) – (c+d) = 0 ?

Rx = ax + bx - cx
Ry = ay + by - cy
R = sqrt(Rx**2 + Ry**2)
print(f"Resultant of vector d is Magnitude = {R}")
angle = atan(Ry/Rx)
if angle < 0:
    angle = 360 + degrees(angle)
else:
    angle = degrees(angle)

print(f"Angle of vector d is : {angle}")
```

#### d) Find the angle between the vector A= 2i -3j+5k and the x, y,and z axes, respectively.

```python
vectA = np.array([2,-3,5])
vectX = np.array([1,0,0])
vectY = np.array([0,1,0])
vectZ = np.array([0,0,1])
```

```python
def dotA(vector1,vector2):
    mag_vector = np.linalg.norm(vector1) * np.linalg.norm(vector2) 
    dotA_ = np.dot(vector1,vector2)
    angle = degrees(acos(dotA_/mag_vector))
    return angle
```

```python
# A = 2i - 3j + 5k with the x-axis
dotA(vectA,vectX)
```

```python
# A = 2i - 3j + 5k with the y-axis
dotA(vectA,vectY)
```

```python
# A = 2i - 3j + 5k with the z-axis
dotA(vectA,vectZ)
```
